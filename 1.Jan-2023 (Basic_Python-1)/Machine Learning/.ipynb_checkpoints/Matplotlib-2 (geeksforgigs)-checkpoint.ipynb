```python
# Combine all above here

x = [1,2,3,4]
y = [2,4,1,6]
plt.plot(x,y)


a = [1,2,3,4,5,6]
b = [0,0.6,2,6,15,8,10,5,16,21]
plt.plot(a)
plt.plot(b,'og')

a = list(range(0,22,3))
plt.plot(a,'or')
plt.xlabel('Day :-->')
plt.ylabel('Temp :-->')

c = [4,2,6,8,15,24,16,10,45]
plt.plot(c,label= '4th Repo',marker = '*')

ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.xticks(list(range(-3,10)))
plt.yticks(list(range(-3,20,3)))

ax.legend(['1st Rap', '2nd Rap','3rd Rap','4th Rap'])

plt.annotate('Temperature V / s Days', xy = (1.01, -2.15))

plt.title('All Features Discussed')
```
